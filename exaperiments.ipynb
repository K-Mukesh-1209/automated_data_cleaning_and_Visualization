```python
## Handling missing values
from sklearn.impute import KNNImputer
def handle_miss_values():
    target_col = 'SalePrice'
    for col in df.columns:
        if col == target_col:
            continue  # Skip target column
        
        missing_ratio = df[col].isnull().mean()  # Percentage of missing values

        if missing_ratio == 0:
            continue  # Skip columns without missing values
        
        if missing_ratio > 0.5:
            print(f" Dropping '{col}' (Too many missing values: {missing_ratio*100:.1f}%)")
            df.drop(columns=[col], inplace=True)
        
        elif df[col].dtype in [np.float64, np.int64]:  # Numeric column
            if df[target_col].dtype in [np.float64, np.int64]:  # Target is numeric
                print(f" Using KNN Imputation for '{col}' (Numerical)")
                imputer = KNNImputer(n_neighbors=5)
                df[[col]] = imputer.fit_transform(df[[col]])
            else:
                print(f" Using Median Imputation for '{col}'")
                df[col].fillna(df[col].median(), inplace=True)
        
        else:  # Categorical column
            print(f" Using Mode Imputation for '{col}'")
            df[col].fillna(df[col].mode()[0], inplace=True)

    return df
    
    
handle_miss_values()
```

```python
def find_oulier():
    threshold=3
    mean=np.mean(df['LotFrontage'])
    std=np.std(df['LotFrontage'])
    for i in df['LotFrontage']:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            return True
    return False 
h=find_oulier()  
h 
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.linear_model import LogisticRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

def determine_missingness_type(df, alpha=0.05):
    """
    Determine if missing data is MCAR, MAR, or MNAR.
    Returns:
        - "MCAR" (Missing Completely at Random)
        - "MAR" (Missing at Random)
        - "MNAR (suspected)" (Missing Not at Random)
    """
    # Check for MCAR using Little's test (requires statsmodels)
    try:
        from statsmodels.stats.missingness import LittleMCARTest
        result = LittleMCARTest(df).test()
        if result.pvalue > alpha:
            return "MCAR"
    except ImportError:
        print("Install 'statsmodels' for Little's MCAR test. Skipping MCAR check.")
    except Exception as e:
        print(f"Skipping MCAR check due to error: {e}")

    # Check for MAR by correlating missingness with other variables
    mar_flag = False
    for col in df.columns[df.isnull().any()]:
        missing_indicator = df[col].isnull().astype(int)
        
        for other_col in df.columns:
            if other_col == col:
                continue  # Skip the same column
            
            # Handle numeric columns with logistic regression
            if pd.api.types.is_numeric_dtype(df[other_col]):
                X = df[[other_col]].dropna()
                y = missing_indicator.loc[X.index]
                
                if len(y.unique()) < 2:
                    continue  # Skip if no variance in missing indicator
                
                try:
                    model = LogisticRegression(max_iter=1000).fit(X, y)
                    if np.abs(model.coef_[0][0]) > 0.1:
                        mar_flag = True
                        break
                except:
                    continue
            # Handle categorical columns with chi-squared test
            else:
                # Create a contingency table
                contingency_table = pd.crosstab(missing_indicator, df[other_col])
                if contingency_table.shape[0] < 2 or contingency_table.shape[1] < 2:
                    continue  # Skip if not enough data
                _, p_value, _, _ = chi2_contingency(contingency_table)
                if p_value < alpha:
                    mar_flag = True
                    break
        if mar_flag:
            break

    if mar_flag:
        return "MAR"
    else:
        return "MNAR (suspected - requires domain knowledge)"
type_data = determine_missingness_type(df, 0.05)
```
